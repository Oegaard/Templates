```python
import yfinance as yf
import datetime as dt
import plotly.express as px


startdate = dt.date.today() - dt.timedelta(days=5*365)
enddate = dt.date.today()

stocks = yf.download('AMBU-B.CO,\
                      BAVA.CO,\
                      CARL-B.CO,\
                      CHR.CO,\
                      COLO-B.CO',
                      start = startdate,
                      end = enddate)['Close']
```

```python
    fig = px.line()
    
    for i,n in enumerate(stocks):
        fig.add_scatter(y= stocks[stocks.columns[i]], name= stocks.columns[i])
        
        
    fig.update_layout(
    title = 'Stock prices'
    ,xaxis_title = 'Dates'
    ,yaxis_title = 'Price in DKK'
    ,font = dict(size = 12)
    )
    
    fig.update_xaxes(rangeslider_visible=True)

    
    fig.update_layout(
    updatemenus=[
        dict(
            #type="buttons",
            direction="down",
            active=0,
            x=0.15,
            y=1.03,
            buttons=list([
                dict(label=stocks.columns[0],
                     method="update",
                     args=[ {"visible": [True, False, False, False, False]},
                            {'showlegend' : True}
                        ]),
                dict(label=stocks.columns[1],
                     method="update",
                     args=[ {"visible": [False, True, False, False, False]},
                            {'showlegend' : True}
                     ]),
                dict(label=stocks.columns[2],
                     method="update",
                     args=[ {"visible": [False, False, True, False, False]},
                            {'showlegend' : True}
                        ]),
                dict(label=stocks.columns[3],
                     method="update",
                     args=[ {"visible": [False, False, False, True, False]},
                            {'showlegend' : True}
                     ]),
                dict(label=stocks.columns[4],
                     method="update",
                     args=[ {"visible": [False, False, False, False, True]},
                            {'showlegend' : True}
                           ]),
                dict(label='All',
                     method="update",
                     args=[ {"visible": [True, True, True, True, True]},
                            {'showlegend' : True}
                           ]),
            ]),
        )
    ]
)
    
    
    fig.show()
    

```
